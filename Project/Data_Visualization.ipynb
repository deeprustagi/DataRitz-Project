```python
# importing packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sb
```

```python
# loading data
df = pd.read_csv(r'C:\DataRitz Project\Person-s-health\Dataset\Dataset.csv',names=['Gender','Height','Weight','Index'])
```

# Bar graph

```python
# divide data
male_df=df[(df['Gender']=='Male')]['Index'].value_counts().to_dict()
male_data=list(dict(sorted(male_df.items())).values())
female_df=df[(df['Gender']=='Female')]['Index'].value_counts().to_dict()
female_data=list(dict(sorted(female_df.items())).values())
```

```python
Index=['Underweight', 'Normalweight', 'Healthy', 'Overweight', 'Obesity']
width=0.40
x=np.arange(len(Index))
plt.bar(x-.2,male_data,width,label='Men',color='blue')
plt.bar(x+.2,female_data,width,label='Women',color='pink')
plt.grid(True)
plt.legend(loc='upper left')
plt.xticks(x,Index,rotation=20,color='b')
plt.yticks(color='b')
plt.title("Counts of Men and Women with Status",color='b',size=15)
plt.show()
```

# Scatter plot

```python
# divide data
male_h=df[(df['Gender']=='Male')]['Height']
male_w=df[(df['Gender']=='Male')]['Weight']
female_h=df[(df['Gender']=='Female')]['Height']
female_w=df[(df['Gender']=='Female')]['Weight']
```

```python
data=((male_h,male_w),(female_h,female_w))
labels=('Male','Female')
color=('r','g')

for data,color,group in zip(data,color,labels):
    x,y=data
    plt.scatter(x,y,alpha=1)    # alpha is intensity of color
plt.title('Height v/s Weight',color='r',size=15)
plt.xlabel('Height in feet',color='g',size=14)
plt.ylabel('Weight in kg',color='g',size=14)
plt.legend(labels=['Male','Female'])
plt.show()
```

# Pie chart

```python
male_data=[34,56,29,89,67]
color=['c','yellowgreen','lightcoral','lightskyblue','gold']
plt.pie(male_data,colors=color,autopct='%1.2f%%',pctdistance=0.5,shadow=True,wedgeprops={"edgecolor":"0",'linewidth': 1,'linestyle': 'solid', 'antialiased': True})
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.legend(Index,title="Index",loc='lower left')
plt.title('Per% of data with Index',color='c',size=15)
plt.show()
```

# Histogram

```python
plt.hist([male_h,female_h],color=['green','orange'],label=['Men','Women'],edgecolor='k',linewidth=1.2,hatch='/',fill=True)
plt.title('Height',color='g',size=15)
plt.xlabel('Height range in feet',color='g',size=14)
plt.ylabel('Number of person',color='g',size=14)
plt.legend()
plt.show()
```

# Distribution plot

```python
# Height distribution
sb.set_style('darkgrid')  
sb.distplot(df['Height'],color='c')
plt.title('Height Distribution',color='r',size=15)
plt.xlabel('Height in feet',color='g',size=14)
plt.ylabel('Mean',color='g',size=14)
plt.show()
```

```python
# Weight distribution
sb.set_style('darkgrid')  
sb.distplot(df['Weight'],color='c')
plt.title('Weight Distribution',color='r',size=15)
plt.xlabel('Weight in kg',color='g',size=14)
plt.ylabel('Mean',color='g',size=14)
plt.show()
```

# Strip plot

```python
sb.stripplot(x='Index',y='Height',data=df)
sb.set_style('darkgrid')  
plt.title('Height Strip',color='r',size=15)
plt.ylabel('Height in feet',color='g',size=14)
plt.xlabel('Index',color='g',size=14)
plt.show()
```

```python
sb.stripplot(x='Index',y='Weight',data=df)
sb.set_style('darkgrid')  
plt.title('Weight Strip',color='r',size=15)
plt.ylabel('Weight in kg',color='g',size=14)
plt.xlabel('Index',color='g',size=14)
plt.show()
```
